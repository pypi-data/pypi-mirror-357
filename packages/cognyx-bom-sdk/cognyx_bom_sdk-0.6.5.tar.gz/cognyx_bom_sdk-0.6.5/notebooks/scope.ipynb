```python
from pyodide.code import eval_code_async

from cognyx_bom_sdk.cognyx.automation_runtime import execute_code

global_variables = {}

global_ignore = [
    "BomClient",
    "__updateCallback",
    "__bom_id",
    "__backendURL",
    "__jwt",
]


# Create a namespace for execution


# Example usage
code = """
def func():
    print("func")

def func1():
    try:
        func()
    except NameError:
        print("func not defined")
    print("func1")

func1()
"""

# Call the execute_code function with your code
result = await execute_code(
    code=code,
    eval_async=eval_code_async,
    global_variables=global_variables,
    global_ignore=global_ignore,
)

for name, value in result["scope"].items():
    print(f"global {name}: {value}")
```
