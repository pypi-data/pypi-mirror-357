```python
%pylab
%load_ext line_profiler
from meshoid import Meshoid
from itertools import product
np.random.seed(42)
N= int(10**6)
x = np.random.rand(N,3)
M = Meshoid(x,des_ngb=100,verbose=True,boxsize=1)

print(np.max(M.D(x) - np.eye(3)))

for i,j in product(range(3),range(3)):
      print(i,j,M.D2(x[:,i]*x[:,j]))
```

```python
x2 = np.random.rand(10**4,3)

# 0'th order
print(np.median(np.abs(M.Reconstruct(x,x2,order=0) - x2)))

# 1st order
print(np.median(np.abs(M.Reconstruct(x,x2,order=1) - x2))) 

# 2nd order
print(np.median(np.abs(M.Reconstruct(x,x2,order=2) - x2))) 
for i,j in product(range(3),range(3)):
    print(i,j,np.median(np.abs(M.Reconstruct(x[:,i]*x[:,j],x2,order=2) -x2[:,i]*x2[:,j])))
```

```python
s = M.Slice(np.sin(x[:,0]*np.pi*2),center=np.array([0.5,0.5,0.5]),size=1.,order=2,res=1024)
np.max(s[:,0] - np.sin(2*np.pi*np.linspace(0, 1, 1024)))
```

```python
f = np.sin(x[:,1]*np.pi*2)
X,Y,Z,fg = M.InterpToGrid(f,center=np.array([0.5,0.5,0.5]),size=1.,order=2,res=512,return_grid=True)
```
